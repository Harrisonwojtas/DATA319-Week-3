##### (a) Scatterplot vs Hexagon plot
A scatterplot is a data visualization that displays data points on a graph, where each point is its own piece of information within the x and y values. A hexagon plot is a visualization that groups data points into hexagonal groups and uses color to illustrate the amount of data points within the group. Scatterplots are best with smaller datasets where seeing the individual pieces of data is important, as they provide more details. Hexagon plots are better for greater datasets where finding and understanding trends is more important.

##### (b) Marginal distribution vs Conditional distribution
Marginal distribution refers to the probability of a single event to occur. Does not take into consideration any other events. It’s obtained by summing or integrating the joint probabilities over all possible values of the other variables. Condidional distribution on the other hand is the probability of an event to occur given another event has already happened.

Marginal distributions gives the probability of one variable by itself, ignoring the others.

##### (c) Probability mass function vs Probability density function
Probability Mass function is the probability that a discrete random variable takes at an exact value. 
Probability density function is the probability that a continous random variable toakes at a specified interval. So the difference between the two probabilities is <i>type of variable<i> and <i>evaluated scope (specific point vs interval)<i>
##### (d) Statistical independence
Probabilistic behavior of each random variable in a set is not influenced by the probablistic behavior of the others. To test if they are statistically independent they need to meet 2 main criteria. For event A and B. 1.) The Prob(AnB) = Prob(A)xProb(B). and 2.)Prob(A|B)=Prob(A).

##### (e) Simpson's paradox
Simpsons paradox is where a common trend appears in several groups of data but is reversed or disepears when the groups are combined. For example (with some made up dataset) looking at motorcycle crashes and helmet use. We may see that the helmet use on motorcycles decrease the likelihood of serious injury when accidents occur compared to not wearing the helmet. This may be because the entire dataset contains both car drivers and motorcyclist. So car drivers don't wear motorcycle helemets and it aids to the misrepresented conclusions of the interpretation of the dataset. This is an example of the simpson's paradox

```python
#####(a) Compute the marginal distributions of X and Y.
import numpy as np
from scipy.stats.contingency import margins

data = np.array([
    [0.00, 0.20],
    [0.15, 0.10],
    [0.25, 0.30]
])
print(data)
print()

# Marginal distributions
#Marginal distribution refers to the probability of a single event to occur. Does not take into consideration any other events. 
#It’s obtained by summing or integrating the joint probabilities over all possible values of the other variables.

#axis 1 is the columns and axis 2 is the rows
marginal_X = np.sum(data, axis=1)  # sum across Y
marginal_Y = np.sum(data, axis=0)  # sum across X

print("Marginal Distribution X:")
print(marginal_X)

print("Marginal Distribution Y:")
print(marginal_Y)
```

```python
##### (b) Compute the expected values and variances of X and Y.

#First need values of X and Y
# Values of X and Y
x_vals = np.array([0, 1, 2])
y_vals = np.array([0, 1])

marginal_X
marginal_Y

# Expected values
#E[X]=0(0.20)+1(0.25)+2(0.55)
EX = np.sum(x_vals * marginal_X)

#E[Y]=0(0.40)+1(0.60)
EY = np.sum(y_vals * marginal_Y)

print(x_vals)
print("Expected Values of X:")
print(EX)
print("Expected Values of Y:")
print(EY)
```

```python
#Now using 
# Expected squares or E[X^2] and E [Y^2]

EX2 = np.sum(x_vals**2 * marginal_X)
EY2 = np.sum(y_vals**2 * marginal_Y)

# Variances
#Variance of random variables must be computed from its marginal distribution using weighted sums
VarX = EX2 - EX**2
VarY = EY2 - EY**2

print("E[X] =", EX)
print("Variance of X is:")
print("Var(X) =", VarX)
print()
print("E[Y] =", EY)
print("Variance of Y is:")
print("Var(Y) =", VarY)
```

```python
####B. Compute sample means, covariances, and correlations of these variables.¶

#Means
mean = LA_data.mean()
print("Sample Means:")
print(mean)

#Covariance
covariance = LA_data.cov()
print("Sample Covariance:")
print(covariance)

#Correlation matrix
correlation = LA_data.corr()

print("Sample correlation:")
print(correlation)

```
